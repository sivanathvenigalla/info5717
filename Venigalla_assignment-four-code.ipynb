```python
import pandas as pd
df = pd.read_csv('assignment_4_property_tax_report_2019.csv')
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
X_axis = df.groupby(by=['YEAR_BUILT'])['PID'].count()
X_axis = X_axis.reset_index()
X_axis = X_axis[(X_axis['YEAR_BUILT']>=1900 ) & (X_axis['YEAR_BUILT']<=2019)]
Y_axis = X_axis['PID']
X_axis = X_axis['YEAR_BUILT']
sns.set_style(rc = {'axes.facecolor': 'lightsteelblue'})
plt.figure(figsize=(20,10))
plt.plot(X_axis,Y_axis)
#plt.xticks(X_axis)
plt.title('A Line Plot to show the number of houses built every year (2019 -2019)')
plt.ylabel('number of houses built')
plt.xlabel('year')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.hist(X_axis, bins=len(X_axis), weights=Y_axis)
plt.title(' Histogram to show the distribution of number of houses built (2019 -2019)')
plt.ylabel('number of houses built')
plt.xlabel('year')
plt.show()
```

```python

req_info = df['ZONE_CATEGORY'].value_counts()
req_info = req_info.head(8)


colors = sns.color_palette("light:#5A9")[0:5]

#plt.set_style(rc = {'axes.facecolor': 'lightsteelblue'})
plt.figure(figsize=(15,15)).patch.set_facecolor('white')
plt.pie(req_info.values, labels=req_info.index,  colors = colors, autopct='%.0f%%')
plt.title('A Pie Chart to present the ratio of Top 8 zone categories of houses (2019 -2019)')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.scatter(X_axis,Y_axis)
#plt.xticks(X_axis)
plt.title('Scatter Plot to show the number of houses (2019 -2019)')
plt.ylabel('number of houses built')
plt.xlabel('year ')
plt.show()
```
