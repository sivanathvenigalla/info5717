#####Every instance of the Student class should have the following attributes:
#####➢ First Name
#####➢ Last Name
#####➢ Middle Name
#####➢ euid, by default euid is blank.
#####➢ GPA, by default the GPA is set to 0
#####➢ Classes taken. This object should store the class number (such as INFO 5717), the semester it was taken it (such as "Fall 2019"), and the grade the student obtained. 
HINT: use a collection data type.

```python
import random
class Student:
    def __init__(self, first_name, last_name, middle_name, euid='', gpa=0.0, classes_taken={}):
        self.first_name = first_name
        self.last_name = last_name
        self.middle_name = middle_name
        self.euid = euid
        self.gpa = gpa
        self.classes_taken = classes_taken
        
    def add_class(self, class_number, semester, grade):
        self.classes_taken[class_number] = {'semester': semester, 'grade': grade}
        return
        
    def print_classes(self):
        for class_number, class_info in self.classes_taken.items():
            print(f"{class_number} taken in {class_info['semester']} with a grade of {class_info['grade']}")
        return
    #---------------------------------------3C---------------------------        
    def generate_euid(self):
        # Get the first three letters of the student's first, middle, and last names
        first_initial = self.first_name[0]
        middle_initial = self.middle_name[0]
        last_initial = self.last_name[0]
        
        # Generate a random four-digit number
        random_number = random.randint(1000, 9999)
        
        # Concatenate the initials and random number to form the EUID
        euid = f"{first_initial}{middle_initial}{last_initial}{random_number}"
        
        # Set the student's EUID attribute to the generated EUID
        self.euid = euid
        return
    #-------------------------------------3d-----------------------------
    def register(self):
        num_classes = int(input("How many classes would you like to register for? "))
        for i in range(num_classes):
            class_number = input("Enter the class number: ")
            semester = input("Enter the semester taken (e.g. Fall 2022): ")
            grade = input("Enter the grade obtained (e.g. A, B+, C): ")
            self.add_class(class_number, semester, grade)
        return
    #-----------------------------------3e--------------------------------
    def calculate_gpa(self):
        total_grade_points = 0
        num_classes = 0
        if len(self.classes_taken) == 0:
            print("No classes taken.")
            return
        for class_info in self.classes_taken.values():
            grade = class_info['grade']
            if grade == 'A':
                grade_points = 4.0
            elif grade == 'A-':
                grade_points = 3.7
            elif grade == 'B+':
                grade_points = 3.3
            elif grade == 'B':
                grade_points = 3.0
            elif grade == 'B-':
                grade_points = 2.7
            elif grade == 'C+':
                grade_points = 2.3
            elif grade == 'C':
                grade_points = 2.0
            elif grade == 'C-':
                grade_points = 1.7
            elif grade == 'D+':
                grade_points = 1.3
            elif grade == 'D':
                grade_points = 1.0
            elif grade == 'F':
                grade_points = 0.0
            else:
                print(f"Invalid grade '{grade}' for class {class_info['class_number']}")
                continue
            
            total_grade_points += grade_points
            num_classes += 1
        self.gpa = total_grade_points / num_classes
        print(f"Student {self.euid}'s GPA: {self.gpa:.2f}")
```
